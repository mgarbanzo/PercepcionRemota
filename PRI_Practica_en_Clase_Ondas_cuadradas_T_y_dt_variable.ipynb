# Comprendiendo FOURIER

Evale las consecuencias espectrales de variar dt y T en una serie de tiempo.

```python
import numpy as np
import pylab as plt
from scipy import signal
```

```python
P = 10 #s
w = 2 * np.pi / P
```

```python
t1 = np.arange(0,100,0.1)
t2 = np.arange(0,100,0.05)
t3 = np.arange(0,200,0.1)
cuadrada1 = signal.square(w*t1)
cuadrada2 = signal.square(w*t2) # Menor dt, Mayor F
cuadrada3 = signal.square(w*t3) # Mayor T, Menor df
```

```python
plt.figure(figsize=(10,2))
plt.plot(t1, cuadrada1, '*r')
plt.plot(t2, cuadrada2, '+g')
plt.plot(t3, cuadrada3, '.b')
```

```python
plt.figure(figsize=(10,2))
plt.plot(t1, cuadrada1, '*r')
plt.plot(t2, cuadrada2, '+g')
plt.plot(t3, cuadrada3, '.b')
plt.xlim(0,2)
```

# Calculando el espectro y sus cambios

```python
FFTcuadrada1 = np.fft.fft(cuadrada1)
FFTcuadrada2 = np.fft.fft(cuadrada2)
FFTcuadrada3 = np.fft.fft(cuadrada3)
XFT1 = np.fft.fftfreq(len(FFTcuadrada1),d=0.1)
XFT2 = np.fft.fftfreq(len(FFTcuadrada2),d=0.05)
XFT3 = np.fft.fftfreq(len(FFTcuadrada3),d=0.1)
```

```python
plt.figure()
plt.plot(XFT1, np.abs(FFTcuadrada1), '.r')
plt.figure()
plt.plot(XFT2, np.abs(FFTcuadrada2), '.g')
plt.figure()
plt.plot(XFT3, np.abs(FFTcuadrada3), '.b')
```

```python
plt.figure()
plt.plot(XFT1, np.abs(FFTcuadrada1), '.r')
plt.plot(XFT2, np.abs(FFTcuadrada2), '.g')
plt.plot(XFT3, np.abs(FFTcuadrada3), '.b')
```
