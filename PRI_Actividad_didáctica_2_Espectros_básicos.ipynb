# ESPECTROS BÁSICOS

Enunciado: Obtenga gráficas de las componentes reales e imaginarias de Fourier de las funciones sin(x), cos(x), sin(x)/x


```python
import numpy as np
import pylab as plt
from scipy import signal
```

```python
T = 10 #s
w = 2 * np.pi / T
```

```python
t = np.arange(0,100,0.11) # Mostrar el error con 0.1 s
C = np.cos(w*t)
S = np.sin(w*t)
Sx = np.sin(w*(t - 50))/(w*(t - 50))
```

```python
plt.figure()
plt.plot(t, Sx, '-r')
plt.plot(t, S, '-g')
plt.plot(t, C, '-b')
```

# Calculando los espectros de las funciones básicas

```python
FTC = np.fft.fft(C)
FTS = np.fft.fft(S)
FTSx = np.fft.fft(Sx)
XFT = np.fft.fftfreq(len(FTC),d=0.11)
```

```python
FTSx
```

# Visualizando espectros de funciones cos(x)

Primero graficamos los valores reales e imaginarios.

```python
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(14,4))
fig.suptitle('Gráficos del coseno')
ax1.plot(XFT,np.real(FTC))
ax2.plot(XFT,np.imag(FTC))
```

Luego graficamos la norma y fase de la FT

```python
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(12,4))
fig.suptitle('Gráficos del cos(x)')
ax1.plot(XFT,np.abs(FTC))
ax2.plot(XFT,np.angle(FTC))
```

# Visualizando espectros de funciones sin(x)

Primero graficamos los valores reales e imaginarios.

```python
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(14,4))
fig.suptitle('Gráficos del sin(x)')
ax1.plot(XFT,np.real(FTS))
ax2.plot(XFT,np.imag(FTS))
```

```python
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(14,4))
fig.suptitle('Gráficos del sin(x)')
ax1.plot(XFT,np.abs(FTS))
ax2.plot(XFT,np.angle(FTS))
```

# Visualizando espectros de funciones sin(x)/x

Primero graficamos los valores reales e imaginarios.

```python
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(14,4))
fig.suptitle('Gráficos del sin(x)/x')
ax1.plot(XFT,np.real(FTSx))
ax2.plot(XFT,np.imag(FTSx))
```

```python
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(14,4))
fig.suptitle('Gráficos del sin(x)')
ax1.plot(XFT,np.abs(FTSx))
ax2.plot(XFT,np.angle(FTSx))
```
