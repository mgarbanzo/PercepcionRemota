# Práctica de contenido espectral

Enunciado: Obtenga gráficas de las componentes reales e imaginarias de Fourier de la funcion cos(x) y de la misma función pero multiplicada por una ventana.


```python
import numpy as np
import pylab as plt
from scipy import signal
```

```python
T = 10 #s
w = 2 * np.pi / T
```

```python
t = np.arange(0,107,0.11) # Mostrar el error con 0.1 s
C = np.cos(w*t)
W = np.hamming(len(t)) # Esta es la ventana que vamos a utilizar
CW = C*W
```

```python
plt.figure()
plt.plot(t, C, '-b')
plt.plot(t, CW, '-r')
```

# Calculando los espectros de las funciones básicas

```python
FTC = np.fft.fft(C)
FTCW = np.fft.fft(CW)
XFT = np.fft.fftfreq(len(FTC),d=0.11)
```

```python
FTCW
```

# Visualizando espectros de funciones cos(x)

Primero graficamos los valores.

```python
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(12,4))
fig.suptitle('Gráficos del cos(x)')
ax1.plot(XFT,np.abs(FTC))
ax1.plot(XFT,np.abs(FTCW))
ax2.plot(XFT,np.angle(FTC))
ax2.plot(XFT,np.angle(FTCW))
```

```python
fig, (ax1) = plt.subplots(1, 1, figsize=(12,4))
fig.suptitle('Gráficos del cos(x)')
ax1.plot(XFT,np.abs(FTC))
ax1.plot(XFT,np.abs(FTCW))
plt.xlim((-1,1))
```

# Contaminación y mapeo espectral

Ahora vamos a utilizar múltiples funciones para comprender la contaminación y el mapeo espectral.

```python
F = 10 #Hz
dF = 0.01 #Hz
```

```python
T1 = 1/2 #Hz
w1 = 2 * np.pi / T1

T2 = 1/0.5 #Hz
w2 = 2 * np.pi / T2

T3 = 1/-3 #Hz
w3 = 2 * np.pi / T3

T4 = 1/4 #Hz
w4 = 2 * np.pi / T4

T5 = 1/7 #Hz
w5 = 2 * np.pi / T5
```

```python
w1, w2, w3, w4, w5
```

```python
t = np.arange(0,1/dF,1/F) # Mostrar el error con 0.1 s
C1 = np.cos(w1*t)
C2 = np.cos(w2*t)
C3 = np.cos(w3*t)
C4 = np.cos(w4*t)
C5 = np.cos(w5*t)
```

```python
plt.figure(figsize=(14,4))
plt.plot(t, C1, '-k')
plt.plot(t, C2, '-r')
plt.plot(t, C3, '-b')
plt.plot(t, C4, '-g')
plt.plot(t, C5, '-m')
```

```python
FTC1 = np.fft.fft(C1)
FTC2 = np.fft.fft(C2)
FTC3 = np.fft.fft(C3)
FTC4 = np.fft.fft(C4)
FTC5 = np.fft.fft(C5)
XFT = np.fft.fftfreq(len(FTC1),d=1/F)
```

```python
plt.figure(figsize=(14,4))
plt.plot(XFT, np.abs(FTC1), '-k')
plt.plot(XFT, np.abs(FTC2), '-r')
plt.plot(XFT, np.abs(FTC3), '*b')
plt.plot(XFT, np.abs(FTC4), '-g')
plt.plot(XFT, np.abs(FTC5), '-c')
```
