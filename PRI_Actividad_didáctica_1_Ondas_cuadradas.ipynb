# SUMAS DE FOURIER

Enunciado: Obtenga una onda cuadrada a partir de la suma de sus componentes de Fourier y obtenga el error de la aproximaci√≥n como funci√≥n del n√∫mero de t√©rminos utilizados

```python
import numpy as np
import pylab as plt
from scipy import signal
```

```python
def CalculaSumaFourier(n):
  aprox = np.zeros_like(t)
  for N in range(1,2*n+1,2):
    Napprox = (1/N) * np.sin(N*w*t)
    aprox = aprox + Napprox
  aprox = 4/np.pi * aprox
  return aprox
```

```python
def CalculaError(y1,y2):
  return np.sum(np.abs(y1-y2))
```

```python
T = 10 #s
w = 2 * np.pi / T
```

```python
t = np.arange(0,100,0.1)
cuadrada = signal.square(w*t)
```

```python
n = 3
```

```python
aprox = CalculaSumaFourier(n)
```

```python
plt.figure()
plt.plot(t, aprox, '-r')
plt.plot(t, cuadrada, '-k')
```

```python
plt.plot(t, np.abs(cuadrada - aprox))
```

```python
Error = np.sum(np.abs(cuadrada - aprox))
print(Error)
```

# Calculando el comportamiento del error

```python
Errores = []
for n in range(1,500):
  Errores.append(CalculaError(cuadrada,CalculaSumaFourier(n)))
```

```python
plt.plot(Errores)
```

```python
plt.plot(np.log(Errores))
```

```python
plt.plot(np.gradient(Errores))
```
